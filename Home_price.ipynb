```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("CSV file/homeprices.csv")
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(df.drop("price", axis="columns"),df.price)
```

```python
import pickle
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

```python
with open('model_pickle','rb') as f:
    mp = pickle.load(f)
```

```python
new_data = pd.DataFrame([[3000, 3, 40]], columns=['area', 'bedrooms', 'age'])
p = mp.predict(new_data)
print(p)
```
